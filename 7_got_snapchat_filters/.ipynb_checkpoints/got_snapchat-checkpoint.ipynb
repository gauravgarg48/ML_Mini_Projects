## Making cool face filters for GOT Characters & yourself!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
# Load Images

img_jb = cv2.imread('Jamie_Before.jpg')
img_ja = cv2.imread('Jamie_After.jpg')

img_g = cv2.imread('glasses.png', cv2.IMREAD_UNCHANGED)
img_m = cv2.imread('mustache.png', cv2.IMREAD_UNCHANGED)

img_ja = cv2.cvtColor(img_ja, cv2.COLOR_BGR2RGB)
img_jb = cv2.cvtColor(img_jb, cv2.COLOR_BGR2RGB)
img_g = cv2.cvtColor(img_g, cv2.COLOR_BGRA2RGBA)
img_m = cv2.cvtColor(img_m, cv2.COLOR_BGRA2RGBA)

print(img_ja.shape, img_jb.shape)
print(img_g.shape, img_m.shape)

# Load haarcascade files for eyes and nose detection

eye_cascade = cv2.CascadeClassifier('frontalEyes35x16.xml')
nose_cascade = cv2.CascadeClassifier('Nose18x15.xml')
```

```python
# Algorithm

def snapchat(img, glasses, mustache):
    eyes = eye_cascade.detectMultiScale(img, 1.3, 3)
    nose = nose_cascade.detectMultiScale(img, 1.3, 3)
    
    # Placing glasses into the image
    (x, y, w, h) = eyes[0]
    mh = int(h*0.2)
    mw = int(w*0.2)
    h = h + 2*mh
    w = w + 2*mw
    x = x - mw
    y = y - mh
    glasses = cv2.resize(glasses, (w, h))
    for i in range(glasses.shape[0]):
        for j in range(glasses.shape[1]):
            if glasses[i, j, 3] > 0:
                img[y+i, x+j, :] = glasses[i, j, :-1]
    
    
    # Placing mustache into the image
    (x, y, w, h) = nose[0]
    y = int(y + 0.5*h)
    w = int(1.1*w)
    h = int(0.8*h)
    mustache = cv2.resize(mustache, (w, h))
    for i in range(mustache.shape[0]):
        for j in range(mustache.shape[1]):
            if mustache[i, j, 3] > 0:
                img[y+i, x+j, :] = mustache[i, j, :-1]
                
    return img
```

```python
img_pred = snapchat(img_jb, img_g, img_m)

plt.figure(figsize = (8, 8))
plt.subplot(1, 2, 1)
plt.imshow(img_pred)
plt.subplot(1, 2, 2)
plt.imshow(img_ja)
plt.show()
```

```python
# Testing

test_img = cv2.imread('test_img.png', cv2.IMREAD_UNCHANGED)
test_img = cv2.cvtColor(test_img, cv2.COLOR_BGRA2RGB)
```

```python
pred_img = test_img.copy('C')
snapchat(pred_img, img_g, img_m)

plt.figure(figsize = (8, 8))
plt.subplot(1, 2, 1)
plt.imshow(test_img)
plt.subplot(1, 2, 2)
plt.imshow(pred_img)
plt.show()
```

```python
pred_img = pred_img.reshape((-1, 3))
df = pd.DataFrame(pred_img, columns = ['Channel 1', 'Channel 2', 'Channel 3'])
df.to_csv('pred_img.csv', index = False)
```
