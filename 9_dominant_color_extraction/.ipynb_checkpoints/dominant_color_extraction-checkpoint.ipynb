## Dominate color extraction for Image Segmentation

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
# Load image

img = cv2.imread('img1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
orig_shape = img.shape

print(orig_shape)

plt.imshow(img)
plt.show()
```

```python
# Extract dominant colors algorithm

def img_clustering(X, k):
    
    km = KMeans(n_clusters=k)
    km.fit(X)
    
    colors = np.array(km.cluster_centers_, dtype='uint8')
    print(colors)
    
    # Plot all dominant colors
    for kx in range(k):
        mat = np.ones((100, 100, 3), dtype='uint8')
        mat[:, :, :] = colors[kx]
        
        plt.subplot(1, k, kx+1)
        plt.axis('off')
        plt.imshow(mat)
    
    plt.show()

    # Image Segmentation
    
    new_img = np.zeros(X.shape, dtype='uint8')
    for ix in range(X.shape[0]):
        new_img[ix] = colors[km.labels_[ix]]
    
    return new_img
```

```python
# Flatten each channel of the image

X = img.reshape((-1, 3))

new_img1 = img_clustering(X, k=2)
new_img1 = new_img1.reshape(orig_shape)

new_img2 = img_clustering(X, k=3)
new_img2 = new_img2.reshape(orig_shape)

new_img3 = img_clustering(X, k=5)
new_img3 = new_img3.reshape(orig_shape)
```

```python
# Plot all images

plt.figure(figsize=(20, 20))

plt.subplot(1, 4, 1)
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(new_img3)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(new_img2)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(new_img1)
plt.axis('off')

plt.show()
```
