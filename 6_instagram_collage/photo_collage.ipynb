## Instagram Style Photo Collage

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Load Images

# Corner images
i0 = cv2.cvtColor(cv2.imread('img0.jpg'), cv2.COLOR_BGR2RGB)
i1 = cv2.cvtColor(cv2.imread('img1.jpg'), cv2.COLOR_BGR2RGB)
i2 = cv2.cvtColor(cv2.imread('img2.jpg'), cv2.COLOR_BGR2RGB)
i3 = cv2.cvtColor(cv2.imread('img3.jpg'), cv2.COLOR_BGR2RGB)

# Middle Image
i4 = cv2.cvtColor(cv2.imread('img4.jpg'), cv2.COLOR_BGR2RGB)

print(i0.shape, i1.shape, i2.shape, i3.shape, i4.shape)
```

```python
w = 50     # width of frame
s1 = 800   # size of corner images
s2 = 600   # size of middle image

i0 = cv2.resize(i0, (s1, s1))
i1 = cv2.resize(i1, (s1, s1))
i2 = cv2.resize(i2, (s1, s1))
i3 = cv2.resize(i3, (s1, s1))
i4 = cv2.resize(i4, (s2, s2))
```

```python
# Create final image matrix

img = np.zeros((2*s1+3*w, 2*s1+3*w, 3), dtype="int64")
img[w:s1+w, w:s1+w] = i0
img[w:s1+w, s1+2*w:2*(s1+w)] = i1
img[s1+2*w:2*(s1+w), w:s1+w] = i2
img[s1+2*w:2*(s1+w), s1+2*w:2*(s1+w)] = i3

x = (2*s1+w-s2)//2
img[x:s2+x+2*w, x:s2+x+2*w] = np.zeros((s2+2*w, s2+2*w, 3))
img[x+w:s2+x+w, x+w:s2+x+w] = i4
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.show()
```
